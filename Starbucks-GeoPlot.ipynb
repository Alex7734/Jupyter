```python
star["color"]="rgb(15,78,250)"
filtRO = star['Country'] == "RO"
filtUS = star['Country'] == "US"
dfRomania = star[filtRO]
dfUSA = star[filtUS]
```

```python
# CHANGE     
dfUsed = dfRomania
# CHANGE      


name = str(dfUsed['Country'].head(1)).split()[1]

data = [dict(
    type='scattergeo',
    lon = dfUsed['Longitude'],
    lat = dfUsed['Latitude'],
    hoverinfo = 'text',
    text = "Country: "+dfUsed["Country"] +"City: " + dfUsed.City + " Street Adress: "+dfUsed["Street Address"]+" Store Name: " + dfUsed['Store Name'],
    mode = 'markers',
    marker=dict(
        sizemode = 'area',
        sizeref = 1,
        size= 10 ,
        line = dict(width=1,color = "rosybrown"),
        color = star["color"],
        symbol = 'cross',
        opacity = 0.7),
)]

layout = dict(
    title = f"Starbucks Locations with {len(dfUsed)} stores in {name}",
    hovermode='closest',
    geo = dict(showframe=True, showland=True, showcoastlines=True, showcountries=True,
               countrywidth=1, projection=dict(type='miller'),
              landcolor = 'rgb(217, 217, 217)',
              subunitwidth=1,
              showlakes = True,
              lakecolor = 'firebrick',
              countrycolor="rgb(5, 5, 5)")
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
print(f'Number of stores:{len(dfUsed)}')
```
